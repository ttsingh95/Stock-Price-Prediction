```python
from scipy import interpolate
import pandas as pd
import csv 
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
df= pd.read_csv(r"C:\Users\Tatiksha\Documents\Art & Science of Forecasting\StockPrice\Private Consumption Growth.csv", encoding= 'utf8')
df.head()
```

```python
df["Num"] = df.index + 1
```

```python
plt.plot(df['Num'],df['Value'],'bo')
```

```python
data = np.array((df['Num'],df['Value']))
```

```python
tck,u = interpolate.splprep(data, s=0)
```

```python
unew = np.arange(0, 1.01, 0.01)
```

```python
out = interpolate.splev(unew, tck)
```

```python
fig = plt.gcf()
fig.set_size_inches(25,8)
plt.plot(out[0], out[1], color='pink')
plt.plot(data[0,:], data[1,:], 'ob')
plt.show()
```
